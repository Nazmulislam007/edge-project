```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'covid_19_data.csv'
df = pd.read_csv(file_path, nrows=500)

fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Death Count
axes[0, 0].bar(df['ObservationDate'], df['Deaths'], color="#f89f58")
axes[0, 0].set_title('Death Count during Covid 19')
axes[0, 0].set_ylabel('Number of Deaths')
axes[0, 0].set_xlabel('Observation Date')
axes[0, 0].tick_params(axis='x', rotation=45)

# Confirmed Cases
axes[0, 1].plot(df['ObservationDate'], df['Confirmed'], color="blue", linestyle="-")
axes[0, 1].set_title('Confirmed Cases during Covid 19')
axes[0, 1].set_ylabel('Number of Confirmed Cases')
axes[0, 1].set_xlabel('Observation Date')
axes[0, 1].tick_params(axis='x', rotation=45)

# Recovered Cases
axes[1, 0].scatter(df['ObservationDate'], df['Recovered'], color="green")
axes[1, 0].set_title('Recovered Cases during Covid 19')
axes[1, 0].set_ylabel('Number of Recovered Cases')
axes[1, 0].set_xlabel('Observation Date')
axes[1, 0].tick_params(axis='x', rotation=45)

# Deaths Confirmed
axes[1, 1].plot(df['Deaths'], df['Confirmed'], color="purple", linestyle="--", marker="o")
axes[1, 1].set_title('Deaths vs Confirmed Cases')
axes[1, 1].set_ylabel('Confirmed Cases')
axes[1, 1].set_xlabel('Deaths')

# Adjust space between graphs
plt.subplots_adjust(hspace=0.5, wspace=0.3)

plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.stackplot( df['Deaths'],df['Country/Region'], alpha=0.8, labels=['Deaths'], color="red")
plt.legend(loc='upper left')  # Removed `reverse=True`
plt.title('Deaths Over Region')
plt.xlabel('Countery/Region')
plt.ylabel('Deaths')
plt.tick_params(axis='x', rotation=45)
```

```python
import matplotlib.pyplot as plt

# Data for the pie chart
labels = [
    "Chlorinated hydrocarbons",
    "Mineral oil",
    "Polycyclic aromatic hydrocarbons",
    "Heavy metals",
    "Phenols",
    "Cyanides",
    "Aromatic hydrocarbons"
]
sizes = [9.6, 28.0, 10.3, 30.2, 5.5, 8.6, 10.4]  # Percentage values
colors = [
    "#003f5c",
    "#58508d",
    "#b17da5",
    "#bc5090",
    "#f89f58",
    "#586482",
    "#ff6361"
]

# Create the pie chart
plt.figure(figsize=(9, 6))
plt.pie(
    sizes,
    labels=labels,
    autopct="%1.1f%%",
    startangle=180,
    colors=colors,
)
plt.title("Distribution of Pollutants", fontsize=14)
plt.legend(loc="upper left", bbox_to_anchor=(1.1, 1), fontsize=10)
plt.tight_layout()

# Show the chart
plt.show()
```
